# 파생변수 생성

```python
# Convert 'date_time' column to datetime format
train['date_time'] = pd.to_datetime(train['date_time'], format='%Y-%m-%d %H')

# Extract month, day, and hour into new columns
train['month'] = train['date_time'].dt.month.astype(float)
train['day'] = train['date_time'].dt.day.astype(float)
train['time'] = train['date_time'].dt.hour.astype(float)

# Display the first few rows to confirm the changes
print(train.head())
```

## 결측치 처리 - test의 '비전기냉방설비운영', '태양광보유' 채우기

```python
# 복사본처리
train_copy = train.copy()
test_copy = test.copy()
```

```python
# 1. 건물별 '비전기냉방설비운영', '태양광보유' 변수 동일 -> train_df의 데이터를 test_df에 채움

building_info = train_copy[['num','비전기냉방설비운영', '태양광보유']].drop_duplicates()

test_copy.drop(columns=['비전기냉방설비운영', '태양광보유'], inplace = True)
test_copy = pd.merge(test, building_info, on= 'num')

test_copy
```

## Train - 전력량 그래프 확인

```python
# 1. 전력량 분포 별 건물 개수 분포
import seaborn as sns
sns.histplot(train_copy.groupby('num')['전력사용량(kWh)'].mean())
```

```python
# 2. 건물별 평균 전력 사용량 계산
mean_power_usage = train.groupby('num')['전력사용량(kWh)'].mean().reset_index()

# 시각화
plt.figure(figsize=(10, 6))
sns.barplot(x='num', y='전력사용량(kWh)', data=mean_power_usage)
plt.title('건물별 평균 전력 사용량')
plt.xlabel('건물 번호')
plt.ylabel('평균 전력 사용량 (kWh)')
plt.show()
```

# Train - 상관행렬

```python
# 전체 데이터_상관계수 행렬 계산
correlation_matrix = train_copy.corr()

# Y와의 상관관계 확인
correlation_with_y = correlation_matrix["전력사용량(kWh)"].sort_values(ascending=False)

# 상관관계 행렬 시각화
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()

# Y와 다른 변수들의 상관관계 출력
correlation_with_y
```
